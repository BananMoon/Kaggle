## 2. 범주형 자료 그래프
- 파이 차트 : 범주형 구성 비율을 원형으로 표현한 그래프

```python
ratio = [34,32,16,18] # 각 영역의 비율
labels = ['Apple', 'Banana','Melon', 'Grapes'] # 각 영역의 이름
explode = [0.05, 0.05, 0.05, 0.05] # 부채꼴이 파이차트의 중심에서 벗어나는 정도
colors = ['silver','gold','whitesmoke','lightgray']

plt.pie(ratio, labels=labels, autopct='%1.f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()

# autopct : 부채꼴 안에 표시될 숫자의 형식 지정
# Startangle : 부채꼴이 그려지는 시작 각도를 설정
# counterclock=False : 시계방향 순서로 부채꼴 영역 표시.
```

```python
# 파이차트 2

colors = ['#ff9999','#ffc000','#8fd9b6','#d395d0']  # hex code 이용하여 다양한 색 시정
wedgeprops={'width':0.7, 'edgecolor':'w', 'linewidth':5}  # wedgeprops로 부채꼴 영역 스타일 설정(width:반지름 비율, edgecolor:테두리색상, linewidth:테두리선의 너비)

plt.pie(ratio, labels=labels, autopct='%1.f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()
```

## 4. 범주형 & 연속형 자료 그래프
### 박스 플롯
- 5개의 요약통계량(최소값,1분위수,중앙값,3분위수,최대값) 제공
- 분포의 대칭성, 이상치 쉽게 파악 가능
- Q1 : 전체 분포 상 왼쪽 1/4에 위치한 값
  Q3 : 전체 분포상에서 왼쪽 3/4에 위치한 값
- Upper fence = Q3 + 1.5 * IQR
  Lower fence = Q1 - 1.5 * IQR
- Fence를 넘어가는 값들은 이상치로 간주하고 점으로 표시

### 산점도
- 서로 다른 연속형 변수의 관계 표현에 쓰인다. (변수들의 관계, 패턴의 유무 파악이 목적)
- 변수는 정수형(int64), 실수형(float64)

```python
# 박스플롯 그리기
# tips 데이터셋 : 고객이 팁을 얼마나 주는지에 대한 데이터 셋
import seaborn as sns
tips = sns.load_dataset('tips') # seaborn 패키지에 내장된 tips라는 DataFrame 사용

plt.boxplot(tips['tip'], sym="bo") # 이상치 마커 모양 sym="bo"

plt.title("Box plot of tip") # title 설정
plt.xticks([1], ['tip']) # x축 이름 설정
plt.show()
```

```python
# Vert를 이용한 박스 플롯
plt.boxplot(tips['tip'],
           sym='rs', # rs= red square
            vert=False # x축과 y축 변경
           )

plt.show()
```

```python
# fig : figure로서, 전체 subplot. n개의 그래프를 담는 전체 사이즈
# ax : axe로서, 전체 중 낱개. n개의 그래프

fig, ax = plt.subplots()
ax.boxplot([tips['total_bill'], tips['tip']], sym="b*")  # b*= blue *
plt.title('Multiple box plots of tips on one Axes')
plt.xticks([1,2],
           ['total_bill','tip']) # x축 이름 설정
```

### ◈ distplot : matplotlib의 hist와 kdeplot 통합하여 나타내는 함수, 분포 & 밀도 확인
- sns.distplot(a=None, bins=None, hist=True, kde=True, rug=False)
- a : Series, DataFrame 형태로 단일 변수 지정
- bins : 히스토그램과 동일
- hist : True일때 히스토그램 출력.
- kde : True일때 커널밀도함수 출력.
- rug : 데이터의 실제 위치를 x축 위에 짧은 선으로 표시

##### 장점 : 데이터의 분포를 직관적으로 파악하기 용이함.
##### 단점: outlier에 의해 영향을 쉽게 받으므로, 전처리가 중요!

```python
import warnings
warnings.filterwarnings('ignore')

# 기본형
sns.distplot(titanic["fare"])
```

```python
# histogram 보지않기 : hist=False
sns.distplot(titanic["fare"],
            hist=False)
```

```python
# rugplot : 데이터 위치를 x축 위에 작은 선분으로 표시
sns.distplot(titanic["fare"], rug=True,
            hist=False)
plt.show()
```

```python
# distplot 그리기
# boxplot을 통해 outlier 확인 => fare가 100이상인 값이 이상치!
sns.boxplot(y="fare", data=titanic)
plt.show()

# outlier 제거 전
sns.distplot(titanic["fare"], hist=False)
plt.show()

# outlier 제거
titanic_low = titanic.query("fare < 100")

# outlier 제거 후
sns.distplot(titanic_low["fare"], hist=False)
plt.show()
```

## ◈ lmplot
- seaborn의 내장 데이터셋인 앤스컴 콰르텟 데이터셋 이용.
- 기술통계량은 비슷하나 분포와 그래프가 매우 다르게 이루어진 4개의 데이터셋.
- 데이터 분석 전 시각화의 중요성 강조하기 위해 만들어진 데이터셋

- sns.lmplot(x='x', y='y', col='범주형변수',hue='범주형변수', data=데이터셋, ci=None, order=1, robust=True/False)
- ci : 1~100 값을 넣어주는데, 회귀추정의 신뢰구간을 표현한다.
- robust : 튀는 값으로 인한 영향을 제거할 수 있다. (True)
- order : 회귀선의 차원 옵션 변경 가능

## ◈ jointplot
- 두 변수 간의 상관관계(scatter plot)와 각 변수의 분포(histogram)을 동시에 보여준다.
- 안쪽은 점 데이터로 분포를 확인하고 바깥쪽은 막대그래프로 밀집도를 확인 => 데이터의 경향을 파악하기 좋음.
- 범주형 데이터에 사용 불가능
- kind\
  scatter(default) : 산점도\
  reg : 회귀선 표시\
  hex : 육각형 그래프\
  kde : 커널 밀집 그래프(등고선처럼 보임)

```python
# 기본형 scatterplot
sns.jointplot(x="total_bill", y="tip", data=tips)

# 회귀선 표시 (kind="reg")
sns.jointplot(x="total_bill", y="tip", data=tips, kind = "reg")

# 육각 그래프 (kind="hex")
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")

# 커널 밀집 그래프 (kde: 등고선)
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde")

plt.show()
```

### ◈ 여러개의 plot 그리기
1. 겹쳐 그리기\
최종적으로 plt.show()를 하기 전 한 셀에 여러개의 그래프를 넣어주기만 하면 됨.

- 서로 다른 종류의 그래프도 겹쳐 그릴 수 있다. ex) scatter plot + line plot 조합 etc..

2-1. 나란히 배치하기 - Subplot\
Subplot : Figure 내에 여러개의 Axes 객체를 만들어 각각의 Axes에 개별적인 plot 생성

2-2. 나란히 배치하기 - FacetGrid\
FacetGrid : 범주형 변수의 조합끼리 그래프를 나누어 그릴 수 있다.
- 한종류의 그래프를 범주형 변수 별로 나누어 볼 때 유리
- Hue를 지정할 떄는 FacetGrid에서 적용하자!


```python
# 1. 겹쳐 그리기
# ex)hue 옵션이 없는 displot에서의 활용
sns.distplot(titanic_low.query("sex =='male'")["fare"], 
             hist=False)
sns.distplot(titanic_low.query("sex =='female'")["fare"], 
             hist=False)
plt.show()
```

```python
# for 구문으로 한번에 해결
fct = titanic_low["sex"].unique()
for f in fct:
    sns.distplot(titanic_low.query("sex == @f")["fare"],  # @f에 male과 female이 들어가게됨.
                hist = False)
plt.show()
```

```python
# 2-(1). 나란히 배치하기 - Subplot
# 양옆으로 2개의 subplot 만들기
fig = plt.figure()

# add_subplot(행 갯수, 열 갯수, 순서)
ax1 = fig.add_subplot(1, 2, 1) # 1행, 2열에 1번째에 배치
ax2 = fig.add_subplot(1, 2, 2) # 1행 2열에 2번째에 배치
sns.boxplot(x = "class", y = "fare", data = titanic, ax = ax1) # ax1에 대해서 boxplot 생성
sns.countplot(x = "class", data=titanic, ax = ax2) # ax2에 대해서 countplot 생성
plt.show()
```

```python
# plt.subplots(행 갯수, 열 갯수) -> 한번에 Figure, Axes 객체를 만들어줘 더 간단함.
fig, ax = plt.subplots(ncols = 2) # 열 갯수 2개
sns.boxplot(x = "class", y = "fare", data=titanic, ax=ax[0])
sns.countplot(x = "class", data=titanic, ax = ax[1]) 
plt.show()
```

```python
# 2-(2). 나란히 배치하기 - FacetGrid
# ex- hue를 지정하면(많은 범주형 변수들로 인해) 더 지저분한 그래프가 됨.
sns.scatterplot(x = "total_bill", y="tip", hue="day", data=tips)
```

```python
# (1)조건에 따라 그리드 나누기
grid = sns.FacetGrid(data = tips, col = "day", hue="day", col_wrap = 2) # col_wrap : 한 행에 총 몇개의 열을 표시할지. 한 행에 2개.

# col, row에 그래프를 생성. (지정된 변수로 할 수 있는 모든 조합들을 생성)

# (2)나눠진 그리드에 그래프 적용하기
grid.map(sns.scatterplot, "total_bill", "tip")

# 그리드변수.map(그래프 종류, 들어갈 변수)
```

#### 동일한 종류의 plot을 변수 별로 보고싶을 때 FacetGrid가 압도적 유리!!!
ex) outlier을 제거한 Titanic_low에서 성별(male/female)에 따른 fare 변수의 분포 확인할 때.

```python
# subplot으로 그리기
fig, ax = plt.subplots(ncols=2)

sns.distplot(titanic_low.query("sex == 'male'")["fare"],
            hist=False, color="steelblue", ax=ax[0])

sns.distplot(titanic_low.query("sex == 'female'")["fare"],
            hist=False, color="darkorange",ax=ax[1])

ax[0].set_title("sex = male")
ax[1].set_title("sex = female")

ax[0].set_ylabel("")
ax[1].set_ylabel("")
ax[1].axes.yaxis.set_visible(False)  # 두번째 그래프의 y축을 보여주지않도록.
plt.show()
```

```python
# FacetGrid로 그리기 : 그래프 크기 알아서 맞춤. 각 그래프에 해당하는 변수를 title로 표시
grid = sns.FacetGrid(data = titanic_low, col="sex", hue="sex") # hue옵션 지정시 그래프별 색상 바꿔줌.
grid.map(sns.distplot, "fare", hist=False)
```
